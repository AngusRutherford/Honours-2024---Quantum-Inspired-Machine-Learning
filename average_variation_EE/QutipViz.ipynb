```python
import numpy as np 
import qutip as qt
from qutip import *
from qutip import Qobj, Bloch
import matplotlib.pyplot as plt
```

# Visualiation Test

```python
x = np.random.uniform(low=0, high=1, size=(100))
plt.plot(x)
```

```python
b = Bloch(view=[0,0], figsize=(5,5))
for i in range(0, 100):
    alpha = np.cos(np.pi/2 * x[i])
    beta = np.sin(np.pi/2 * x[i])
    psi = Qobj([[alpha], [beta]])
    b.add_states(psi, kind='point', alpha=0.3)
```

```python
b.show()
```

# Plot Site 2

```python
def plot_vec(vec):
    x1 = vec[0]
    x2 = vec[1]
    psi = Qobj([[x1], [x2]])
    return psi 
```

```python
v1 = [-0.9367914637324803,
  0.3498882013986141]
v2 = [0.3498882013986141,
 0.9367914637324803]
```

```python
b = Bloch(view=[0,0], figsize=(5,5))
b.add_states(plot_vec(v1))
b.add_states(plot_vec(v2))
b.show()
```
